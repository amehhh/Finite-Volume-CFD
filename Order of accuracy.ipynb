## Order of Accuracy Calculation

### Mesh-80×40:
#### L2 norm error=5.59408566936024×10e−5

### Mesh-40×20:
#### L2 norm error=0.00022309654926

### Mesh-20×10:
#### L2 norm error=0.000883298812111

### Mesh-10×5:
#### L2 norm error=0.003992229611405

### Grid Spacings (Δx):
#### For 80×40 Δx=2/80=0.025

#### For 40×20 Δx=2/40=0.05

#### For 20×10 Δx=2/20=0.1

#### For 10×5 Δx=2/10=0.2

## Plot for Mesh 80 x 40 , 40 x 20 , 20 x 10

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot
delta_x = [0.025, 0.05, 0.1]  # Grid spacings
L2_errors = [5.59408566936024e-5, 0.00022309654926, 0.000883298812111]  # L2 norm errors

#Log-Log plot
plt.figure(figsize=(10, 8))
plt.loglog(delta_x, L2_errors, marker='o', color='black', linestyle='-', linewidth=1, label='L2 Norm Error Line') 

mesh_labels = ["Mesh 80,40", "Mesh 40,20", "Mesh 20,10"]
colors = ['red', 'blue', 'green']
for i in range(len(delta_x)):
    plt.loglog(delta_x[i], L2_errors[i], marker='o', color=colors[i], markersize=8, label=mesh_labels[i])

# Calculate the slope
log_delta_x = np.log10(delta_x)
log_L2_errors = np.log10(L2_errors)
slope, intercept = np.polyfit(log_delta_x, log_L2_errors, 1) # Linear fit to log-log data

# Annotation on the slope on the graph
slope_text = f"Slope = {slope:.2f} ≈ 2  (2nd Order Accurate!) "
plt.text(0.04, 1e-4, slope_text, fontsize=12, color='blue', bbox=dict(facecolor='white', alpha=0.8))


# Labels, title, and grid
plt.xlabel('Log(Grid Spacing Δx)', fontsize=12)
plt.ylabel('Log(L2 Norm Error)', fontsize=12)
plt.title('Log-Log Plot of L2 Norm Error vs Grid Spacing', fontsize=14)
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
